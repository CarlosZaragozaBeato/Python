```python
python = (85,67,23,98)
java = (50,67,89,14)
networking = (60,20,56,22)
machine_learning = (88,23,40,87)
```

```python
people = ["CARLOA","GAUCHO", "CARLAO", "FENOMENO"]
```

```python
index = np.arange(4)
```

```python
plt.bar(index, python, width=0.2, label="Python")
plt.bar(index+0.2, java, width=0.2, label="Java")
plt.bar(index+0.4, networking, width=0.2, label="networking")
plt.bar(index+0.6, machine_learning, width=0.2, label="machine_learning")

plt.title('It Levels')
plt.xlabel('Person')
plt.ylabel('skill')
plt.xticks(index+0.2, people)
plt.legend(loc='upper right')
plt.ylim(0,150)
```

```python
nationalities = ['American', 'German', 'french','other']
students = [60,23,21,34]
```

```python
plt.pie(students, labels=nationalities, autopct='%.2f%%', shadow=True)
```

```python
explode = [0,0,0.1,0]
plt.pie(students, labels=nationalities, autopct='%.2f%%', shadow=True,explode=explode)
```

```python
pairs = zip(students, nationalities)
pairs = sorted(pairs)
students, nationalities = zip(*pairs)
plt.pie(students, labels=nationalities, autopct='%.2f%%', shadow=True,counterclock=False,startangle=90)
plt.title('student Nationality')
```

```python
z = np.linspace(0,30,100)
x = np.sin(z)
y = np.cos(z)
ax = plt.axes(projection="3d")
ax.plot3D(x,y,z)
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y) 
Z = np.sin(np.sqrt(X**2+Y**2))
ax.plot_surface(X,Y,Z)
```
